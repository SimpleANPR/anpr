```python
from anpr.datasets.open_alpr import OpenALPRDataset
from anpr.generic.grayscale import GrayScale
from anpr.detection.wavelet_detector import WaveletDetector

ds = OpenALPRDataset()
img = GrayScale().process(ds.image_at(26).image)
```

```python
from matplotlib import pyplot as plt

plt.imshow(img,cmap='gray')
```

```python
detector = WaveletDetector()

result = detector.detect(img)
```

```python
plt.imshow(result.plate_image,cmap='gray')
```

```python
#o primeiro candidato não foi o correto, mas o terceiro estava certo
plt.imshow(result.extras.candidates[2],cmap='gray')
```
