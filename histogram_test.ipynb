```python
from anpr.generic.brightness import Brightness
from anpr.generic.contrast import Contrast
from anpr.generic.histogram_normalization import HistogramNormalization
from anpr.generic.grayscale import GrayScale
from anpr.datasets.open_alpr import OpenALPRDataset

from matplotlib import pyplot as plt
```

```python
ds = OpenALPRDataset()
img = ds.image_at(20).image
plt.imshow(img)
```

```python
plt.imshow(Brightness(50).process(img))
```

```python
plt.imshow(Brightness(-50).process(img))
```

```python
plt.imshow(Contrast(2.4).process(img))
```

```python
plt.imshow(Contrast(0.5).process(img))
```

```python
plt.imshow(HistogramNormalization().process(img))
```

```python
plt.imshow(GrayScale().process(img), cmap='gray')
```

```python
plt.imshow(HistogramNormalization().process(GrayScale().process(img)), cmap='gray')
```
