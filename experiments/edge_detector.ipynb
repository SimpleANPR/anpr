```python
from matplotlib import pyplot as plt

from anpr.datasets.open_alpr import OpenALPRDataset
from anpr.generic.grayscale import GrayScale
from anpr.generic.laplacian import Laplacian
from anpr.generic.sobel import Sobel
from anpr.generic.prewitt import Prewitt
from anpr.detection.edge_detector import EdgeDetector

ds = OpenALPRDataset()
img = ds.image_at(105).image
```

```python
plt.imshow(img)
```

```python
detector = EdgeDetector(high_pass=Prewitt())

result = detector.detect(img)

result.found_plate
```

```python
plt.imshow(result.plate_image, cmap='gray')
```

```python
#o primeiro candidato não foi o correto, mas o terceiro estava certo
plt.imshow(result.extras.candidates_img[2])
```

```python
img_2 = ds.image_at(90).image
result_2 = detector.detect(img_2)
```

```python
plt.imshow(img_2)
```

```python
import matplotlib.patches as patches

fig, ax = plt.subplots()
ax.imshow(img_2)

if result_2.n_candidates > 0:
    for c in result_2.extras.candidates:
        x, y, w, h = c
        ax.add_patch(patches.Rectangle((x, y), w, h, 
                                    linewidth=1, 
                                    edgecolor='r', 
                                    facecolor='none'))
```
