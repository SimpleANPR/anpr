```python
from anpr.generic.grayscale import GrayScale
from anpr.datasets.open_alpr import OpenALPRDataset
from matplotlib import pyplot as plt
import matplotlib.patches as patches
import cv2
```

```python
ds = OpenALPRDataset()
```

```python
img = ds.image_at(100)
print(img.plate_text)

fig, ax = plt.subplots()
ax.imshow(img.image)
x, y, w, h = img.plate_rect
ax.add_patch(patches.Rectangle((x, y), w, h, 
                               linewidth=1, 
                               edgecolor='r', 
                               facecolor='none'))
```

```python
plt.imshow(GrayScale(cv2.COLOR_RGB2GRAY).process(img.image),
           cmap='gray')
```
