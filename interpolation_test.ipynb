```python
import math
from anpr.interpolation.lanczos import Lanczos
from anpr.interpolation.bilinear import Bilinear

from anpr.datasets.open_alpr import OpenALPRDataset

from matplotlib import pyplot as plt
```

```python
ds = OpenALPRDataset()
img = ds.image_at(20).image
target_res = (200, 200)
imgs = {
    'Original': img,
    'Lanczos': Lanczos(target_res).resize(img),
    'Bilinear': Bilinear(target_res).resize(img)
}


cols = 2
rows = math.ceil(len(imgs) / cols)
fig, axs = plt.subplots(rows, cols,
                        figsize=(4 * cols, 4 * rows),
                        facecolor='w',
                        edgecolor='k')
axs = axs.ravel()

for i, (k, v) in enumerate(imgs.items()):
  axs[i].set_title(k)
  axs[i].imshow(v)
```
